# Chemical Sample Table (CST) Generator

This notebook demonstrates the use of the `create_cst` function to create a Chemical Sample Table from FAIRFluids JSON files.

The function:
- Extracts IUPAC names and CAS Registry Numbers from compounds
- Automatically fetches missing data from PubChem
- Creates an XLSX file with the collected information

```python
import sys
from pathlib import Path

# Add the main directory to the Python path
project_root = Path().resolve().parent.parent
sys.path.insert(0, str(project_root))

from fairfluids.core.create_cst import create_cst
import pandas as pd
```

## Example 1: Process a single JSON file

Processes a single JSON file and creates a Chemical Sample Table.

```python
# Path to the JSON file
json_file = "spera_et_al_fpe_592_2025_114324.json"
output_file = "chemical_sample_table_spera.xlsx"

# Create the Chemical Sample Table
df = create_cst(json_file, output_file)

# Display the first rows
print("\nFirst rows of the Chemical Sample Table:")
print(df.head(10))
```

## Example 2: Combine multiple JSON files

Processes multiple JSON files simultaneously and combines all compounds in a single table.

```python
# List of JSON files
json_files = [
    "spera_et_al_fpe_592_2025_114324.json",
    "Xu_Glycerol_Water.json"
]

output_file = "chemical_sample_table_combined.xlsx"

# Create the combined Chemical Sample Table
df_combined = create_cst(json_files, output_file)

# Display statistics
print("\nStatistics of the combined table:")
print(f"Total number of compounds: {len(df_combined)}")
print(f"\nCompounds with IUPAC Name: {df_combined['IUPAC Name'].notna().sum()}")
print(f"Compounds with CAS Registry Number: {df_combined['CAS Registry Number'].notna().sum()}")
print(f"\nCompounds without IUPAC Name: {df_combined['IUPAC Name'].isna().sum()}")
print(f"Compounds without CAS Registry Number: {df_combined['CAS Registry Number'].isna().sum()}")

# Display the complete table
print("\n\nComplete Chemical Sample Table:")
df_combined
```

## Example 3: With additional fields

Creates an extended table with additional information such as Common Name and PubChem ID.

```python
# Create extended table with additional fields
output_file_extended = "chemical_sample_table_extended.xlsx"

df_extended = create_cst(
    json_files,
    output_file_extended,
    include_additional_fields=True
)

# Display the extended table
print("Extended Chemical Sample Table with additional fields:")
df_extended
```

## Example 4: Data analysis

Analyze the created data and display details about missing information.

```python
# Analyze the data
print("=== Data Analysis ===\n")

# Display compounds without IUPAC Name
missing_iupac = df_extended[df_extended['IUPAC Name'].isna()]
if len(missing_iupac) > 0:
    print(f"Compounds without IUPAC Name ({len(missing_iupac)}):")
    print(missing_iupac[['Compound ID', 'Common Name', 'PubChem ID']])
else:
    print("✓ All compounds have an IUPAC name")

print("\n" + "="*50 + "\n")

# Display compounds without CAS Registry Number
missing_cas = df_extended[df_extended['CAS Registry Number'].isna()]
if len(missing_cas) > 0:
    print(f"Compounds without CAS Registry Number ({len(missing_cas)}):")
    print(missing_cas[['Compound ID', 'Common Name', 'IUPAC Name', 'PubChem ID']])
else:
    print("✓ All compounds have a CAS Registry Number")

print("\n" + "="*50 + "\n")

# Display complete entries
complete = df_extended[
    df_extended['IUPAC Name'].notna() & 
    df_extended['CAS Registry Number'].notna()
]
print(f"Complete entries (with IUPAC Name and CAS Registry Number): {len(complete)}")
if len(complete) > 0:
    print("\nExamples of complete entries:")
    print(complete.head())
```
