### Understanding the FAIRFluids Document Structure

Before we proceed to add citation information, it's important to understand the fundamental structure of a FAIRFluids document. The `FAIRFluidsDocument` class serves as the root container for all fluid property data and metadata. It follows the FAIR (Findable, Accessible, Interoperable, Reusable) principles by providing:

1. **Structured Metadata**: Every document includes version information to track changes and ensure reproducibility
2. **Citation Management**: Proper attribution and referencing capabilities for scientific data
3. **Compound Registry**: A comprehensive database of chemical compounds with multiple identifiers
4. **Fluid Systems**: Experimental and computational data organized by fluid systems (pure substances or mixtures)
5. **Validation Framework**: Built-in data validation to ensure data quality and consistency

The version system allows researchers to track document evolution, maintain data provenance, and ensure that different versions of the same dataset can be properly distinguished. This is crucial for scientific reproducibility and data management in collaborative research environments.


```python
# Create a new document
doc = FAIRFluidsDocument(
    version=Version(versionMajor=1, versionMinor=0)
)

print(f"Created document with version {doc.version.versionMajor}.{doc.version.versionMinor}")
```

### The Importance of Proper Citation in Scientific Data

Citation information is crucial for maintaining scientific integrity and enabling proper attribution in FAIRFluids documents. The `Citation` class provides comprehensive metadata about the source of the data, including:

**Literature Types**: The system supports various publication types including:
- **Journal Articles**: Peer-reviewed research papers
- **Books**: Monographs and reference works
- **Conference Proceedings**: Presentations and abstracts
- **Theses**: Doctoral and master's dissertations
- **Reports**: Technical reports and white papers
- **Patents**: Intellectual property documentation

**Author Management**: Each citation can include multiple authors with detailed information:
- **Names**: Given and family names for proper identification
- **Affiliations**: Institutional connections for context
- **ORCID IDs**: Unique researcher identifiers for disambiguation
- **Contact Information**: Email addresses for correspondence

**Publication Details**: Comprehensive publication metadata including:
- **DOI**: Digital Object Identifiers for permanent linking
- **Volume/Issue**: Traditional journal citation elements
- **Page Numbers**: Specific location within the publication
- **Publication Year**: Temporal context for the data
- **URLs**: Direct links to online resources

This structured approach ensures that every piece of data can be traced back to its original source, facilitating proper scientific attribution and enabling readers to access the full context of the research.

### Building a Comprehensive Compound Database

The compound registry is the foundation of any FAIRFluids document, providing unique identification and structural information for each chemical species. The `Compound` class offers multiple identification systems to ensure maximum interoperability:

**Primary Identifiers**:
- **Compound ID**: Internal unique identifier within the document
- **PubChem ID**: Integration with the world's largest chemical database
- **Common Name**: Human-readable chemical names (e.g., "Water", "Ethanol")

**Structural Representations**:
- **IUPAC Names**: Standardized chemical nomenclature following international conventions
- **SMILES Codes**: Linear notation for molecular structure representation
- **SELFIES**: Robust molecular string representations for machine learning applications
- **InChI/InChI Key**: International Chemical Identifier for unique molecular representation

**Physical Properties**:
- **Molar Weight**: Molecular mass in g/mol for stoichiometric calculations
- **Sigma Profile**: Surface charge distribution for ionic liquid characterization

**Best Practices for Compound Registration**:
1. **Always include PubChem ID** when available for database integration
2. **Provide multiple identifiers** to ensure cross-platform compatibility
3. **Use standardized nomenclature** (IUPAC names) for unambiguous identification
4. **Include structural representations** for computational applications
5. **Validate identifiers** against authoritative databases when possible

This comprehensive approach ensures that compounds can be uniquely identified across different systems and databases, facilitating data sharing and integration in collaborative research environments.


```python
import fairfluids
from fairfluids import FAIRFluidsDocument, Version, Citation, LitType

# Create a new document
doc = FAIRFluidsDocument(
    version=Version(versionMajor=1, versionMinor=0)
)

# Add citation information using only the allowed fields/blocks
doc.citation = Citation(
    litType=LitType.JOURNAL,
    title="Thermodynamic Properties of Ionic Liquids",
    pub_name="Journal of Chemical Thermodynamics",
    publication_year="2024",
    lit_volume_num="45",
    page="123"
)
doc.citation.add_to_author(given_name="John", family_name="Doe")
doc.citation.add_to_author(given_name="Jane", family_name="Smith")

print(f"Added citation with type: {doc.citation.litType}")
print(f"Title: {doc.citation.title}")
print(f"Publication Name: {doc.citation.pub_name}")
print(f"Publication Year: {doc.citation.publication_year}")
print(f"Volume: {doc.citation.lit_volume_num}")
print(f"Page: {doc.citation.page}")
print(f"Authors: {len(doc.citation.author)} authors")
```

### Organizing Experimental Data Through Fluid Systems

The FAIRFluids framework organizes experimental and computational data through the concept of "fluid systems" - containers that group related measurements, properties, and parameters. This hierarchical structure provides several advantages:

**Fluid System Architecture**:
- **Compound References**: Each fluid system references one or more compounds from the compound registry
- **Property Definitions**: Physical quantities being measured (density, viscosity, thermal conductivity, etc.)
- **Parameter Specifications**: Experimental conditions (temperature, pressure, composition)
- **Measurement Collections**: Actual data values with uncertainties and methodological information

**Property Categories**: The system supports a comprehensive range of fluid properties:
- **Thermodynamic Properties**: Density, heat capacity, compressibility, critical properties
- **Transport Properties**: Viscosity, thermal conductivity, diffusion coefficients
- **Phase Behavior**: Boiling points, melting points, vapor pressure, surface tension
- **Thermodynamic Functions**: Enthalpy, entropy, Gibbs free energy, fugacity coefficients

**Parameter Types**: Experimental conditions are systematically categorized:
- **Thermodynamic State**: Temperature, pressure, volume
- **Composition**: Mole fractions, mass fractions, concentrations
- **System Constraints**: Fixed ratios, constant properties
- **Environmental**: Frequency, wavelength, ionic strength

**Measurement Methodology**: Each measurement includes:
- **Method Classification**: Experimental, calculated, simulated, or literature-derived
- **Uncertainty Quantification**: Standard deviations, confidence intervals
- **Source Attribution**: DOI references, experimental protocols
- **Quality Indicators**: Validation flags, precision estimates

This structured approach ensures that all experimental context is preserved, enabling proper interpretation and reproduction of scientific results.

### Modeling Complex Fluid Systems: Binary Mixtures and Beyond

Real-world applications often involve complex fluid systems with multiple components. The FAIRFluids framework handles these through sophisticated composition modeling:

**Binary and Multicomponent Systems**:
- **Pure Substances**: Single-component fluids with well-defined properties
- **Binary Mixtures**: Two-component systems with composition-dependent properties
- **Multicomponent Mixtures**: Complex systems with three or more components
- **Ionic Liquids**: Specialized systems with ionic character and unique behavior

**Composition Parameterization**:
- **Mole Fractions**: Molar composition ratios (most common in thermodynamics)
- **Mass Fractions**: Weight-based composition ratios
- **Volume Fractions**: Volume-based composition ratios
- **Concentration Units**: Molarity, molality, mass concentration
- **Activity Coefficients**: Thermodynamic activity relationships

**Associated Compound Parameters**: The framework allows parameters to be associated with specific compounds:
- **Component-Specific Properties**: Properties that depend on individual components
- **Partial Properties**: Component contributions to mixture properties
- **Excess Properties**: Deviations from ideal mixing behavior
- **Activity Coefficients**: Individual component activities

**Advanced Composition Features**:
- **Solvent-Solute Relationships**: Distinguishing between solvent and solute components
- **Binary Solvent Systems**: Complex solvent mixtures with multiple components
- **Concentration Gradients**: Systems with varying composition profiles
- **Phase Separation**: Systems exhibiting liquid-liquid or liquid-solid equilibria

**Best Practices for Mixture Modeling**:
1. **Always specify composition parameters** for multicomponent systems
2. **Use consistent composition units** throughout the dataset
3. **Include associated compound references** for component-specific parameters
4. **Validate composition constraints** (e.g., mole fractions sum to unity)
5. **Document composition ranges** and phase behavior limits

This comprehensive approach enables accurate modeling of complex fluid systems encountered in industrial applications, environmental studies, and advanced research scenarios.

### Ensuring Data Quality Through Validation

Data validation is a critical step in maintaining scientific integrity and ensuring that FAIRFluids documents meet quality standards. The framework provides comprehensive validation capabilities:

**Structural Validation**:
- **Schema Compliance**: Ensures all required fields are present and properly formatted
- **Type Checking**: Validates data types match expected formats (strings, numbers, enums)
- **Reference Integrity**: Verifies that compound references exist in the compound registry
- **Constraint Validation**: Checks that values fall within reasonable physical ranges

**Content Validation**:
- **Physical Consistency**: Validates that property values are physically reasonable
- **Unit Consistency**: Ensures parameter and property units are compatible
- **Composition Constraints**: Verifies that mole fractions sum to unity in mixtures
- **Temperature-Pressure Ranges**: Checks that conditions are within valid phase regions

**Metadata Validation**:
- **Citation Completeness**: Ensures all required citation fields are populated
- **Identifier Uniqueness**: Verifies that compound IDs and measurement IDs are unique
- **DOI Validation**: Checks DOI format and accessibility when provided
- **Author Information**: Validates author names and contact information

**Quality Assurance Features**:
- **Automated Checks**: Built-in validation rules for common data quality issues
- **Error Reporting**: Detailed error messages with suggestions for corrections
- **Warning System**: Non-critical issues that don't prevent validation but indicate potential problems
- **Completeness Scoring**: Quantitative assessment of data completeness

**Validation Best Practices**:
1. **Validate Early and Often**: Check data quality during input rather than at the end
2. **Use Consistent Units**: Standardize units across all measurements
3. **Document Uncertainties**: Always include uncertainty estimates when available
4. **Cross-Reference Data**: Verify consistency between related measurements
5. **Maintain Audit Trails**: Keep records of validation results and corrections

This rigorous validation process ensures that FAIRFluids documents maintain the highest standards of scientific data quality and reliability.

The JSON export functionality is crucial for data sharing, archival, and integration with other scientific tools. FAIRFluids documents are serialized using a structured JSON format that maintains all metadata and relationships:

**Serialization Features**:

- **Type Information**: Data types are preserved for proper reconstruction
- **Validation Compatibility**: Exported data can be validated against the same schema

**Interoperability Considerations**:
- **Standard Format**: JSON is widely supported across scientific computing platforms
- **API Integration**: Easy integration with REST APIs and web services

- **Cloud Platforms**: Native support in cloud storage and processing services

**Data Archival Best Practices**:

This robust export system ensures that FAIRFluids data can be seamlessly integrated into existing scientific workflows and shared across different research groups and institutions.


```python
# Add experimental data by creating a fluid with measurements
from fairfluids import Fluid, Property, Parameter, Measurement, PropertyValue, ParameterValue, Properties, Parameters, Method
from fairfluids.core.lib import Sample

# Create a fluid for water measurements
water_fluid = doc.add_to_fluid(
    compounds=["1"],  # Reference to water compound
    property=[],
    parameter=[],
    sample=Sample(measurement=[])
)

# Add properties for density and viscosity
density_prop = water_fluid.add_to_property(
    propertyID="density_prop",
    properties=Properties.DENSITY,
    unit=None
)

viscosity_prop = water_fluid.add_to_property(
    propertyID="viscosity_prop",
    properties=Properties.VISCOSITY,
    unit=None
)

# Add parameters for temperature and pressure
temp_param = water_fluid.add_to_parameter(
    parameterID="temp_param",
    parameters=Parameters.TEMPERATURE,
    unit=None
)

pressure_param = water_fluid.add_to_parameter(
    parameterID="pressure_param",
    parameters=Parameters.PRESSURE,
    unit=None
)

# Add measurement for water
water_measurement = water_fluid.sample.add_to_measurement(
    measurement_id="exp_001",
    source_doi=None,
    propertyValue=[],
    parameterValue=[],
    method=Method.MEASURED,
    method_description="Experimental measurement"
)

# Add property values
water_measurement.add_to_propertyValue(
    propertyID="density_prop",
    propValue=997.0,
    uncertainty=None
)

water_measurement.add_to_propertyValue(
    propertyID="viscosity_prop",
    propValue=0.00089,
    uncertainty=None
)

# Add parameter values
water_measurement.add_to_parameterValue(
    parameterID="temp_param",
    paramValue=298.15,
    uncertainty=None
)

water_measurement.add_to_parameterValue(
    parameterID="pressure_param",
    paramValue=101325.0,  # Pressure in Pa
    uncertainty=None
)

print(f"Added fluid with {len(water_fluid.sample.measurement)} measurements")
print(f"  - Measurement ID: {water_fluid.sample.measurement[0].measurement_id}")
print(f"  - Properties: {len(water_fluid.property)}")
print(f"  - Parameters: {len(water_fluid.parameter)}")
```

## Step 5: Add mixture information

Create a binary mixture of choline chloride and glycerol.

```python
# Create a binary fluid system (choline chloride + glycerol mixture)
binary_fluid = doc.add_to_fluid(
    compounds=["2", "3"],  # Reference to choline chloride and glycerol compounds
    property=[],
    parameter=[],
    sample=Sample(measurement=[])
)

# Add properties for the binary mixture
density_prop_mix = binary_fluid.add_to_property(
    propertyID="density_prop_mix",
    properties=Properties.DENSITY,
    unit=None
)

viscosity_prop_mix = binary_fluid.add_to_property(
    propertyID="viscosity_prop_mix",
    properties=Properties.VISCOSITY,
    unit=None
)

# Add parameters for temperature, pressure, and composition
temp_param_mix = binary_fluid.add_to_parameter(
    parameterID="temp_param_mix",
    parameters=Parameters.TEMPERATURE,
    unit=None
)

pressure_param_mix = binary_fluid.add_to_parameter(
    parameterID="pressure_param_mix",
    parameters=Parameters.PRESSURE,
    unit=None
)

# Add mole fraction parameter for choline chloride
mole_fraction_param = binary_fluid.add_to_parameter(
    parameterID="mole_fraction_choline",
    parameters=Parameters.MOLE_FRACTION,
    unit=None,
    associated_compounds=["2"]  # Associated with choline chloride
)

# Add measurement for the binary mixture
mix_measurement = binary_fluid.sample.add_to_measurement(
    measurement_id="mix_001",
    source_doi=None,
    propertyValue=[],
    parameterValue=[],
    method=Method.MEASURED,
    method_description="Experimental measurement of binary mixture"
)

# Add property values for the mixture
mix_measurement.add_to_propertyValue(
    propertyID="density_prop_mix",
    propValue=1200.0,  # Example density for the mixture
    uncertainty=None
)

mix_measurement.add_to_propertyValue(
    propertyID="viscosity_prop_mix",
    propValue=0.0015,  # Example viscosity for the mixture
    uncertainty=None
)

# Add parameter values
mix_measurement.add_to_parameterValue(
    parameterID="temp_param_mix",
    paramValue=298.15,
    uncertainty=None
)

mix_measurement.add_to_parameterValue(
    parameterID="pressure_param_mix",
    paramValue=101325.0,
    uncertainty=None
)

mix_measurement.add_to_parameterValue(
    parameterID="mole_fraction_choline",
    paramValue=0.5,  # 50% mole fraction of choline chloride
    uncertainty=None
)

print(f"Created binary fluid system with compounds: {binary_fluid.compounds}")
print(f"  - Properties: {len(binary_fluid.property)}")
print(f"  - Parameters: {len(binary_fluid.parameter)}")
print(f"  - Measurements: {len(binary_fluid.sample.measurement)}")
```

## Step 6: Validate the document

Check if the document structure is valid.

```python
# Validate the document structure
try:
    # This will raise an error if the document is invalid
    doc.model_validate(doc.model_dump())
    print(" Document validation successful!")

    # Print document summary
    print(f"\n Document Summary:")
    print(f"  - Version: {doc.version.versionMajor}.{doc.version.versionMinor}")
    print(f"  - Compounds: {len(doc.compound)}")
    print(f"  - Fluids: {len(doc.fluid)}")

    # Count total measurements across all fluids
    total_measurements = sum(len(fluid.sample.measurement) for fluid in doc.fluid)
    print(f"  - Total Measurements: {total_measurements}")

except Exception as e:
    print(f"Document validation failed: {e}")
```

## Step 7: Save to JSON

Export the document to a JSON file for persistence and sharing.

```python
# Save to JSON
with open('fairfluids_model.json', 'w') as f:
    f.write(doc.model_dump_json(indent=2))

print("Document saved to 'fairfluids_model.json'")
print(f"File size: {len(doc.model_dump_json())} characters")
```

## Step 8: Load and verify the saved document

Demonstrate loading the saved document back.

```python
# Load the saved document
with open('fairfluids_model.json', 'r') as f:
    loaded_data = json.load(f)

# Create a new document from the loaded data
loaded_doc = FAIRFluidsDocument.model_validate(loaded_data)

print("Successfully loaded the saved document!")
print(f"Loaded document has {len(loaded_doc.compound)} compounds")
print(f"First compound: {loaded_doc.compound[0].commonName}")
```

## Step 9: Advanced features - Search and filter

Demonstrate searching and filtering capabilities.

```python
# Search for compounds by name
def find_compound_by_name(doc, name_fragment):
    """Find compounds containing the given name fragment."""
    matches = []
    for comp in doc.compound:
        if name_fragment.lower() in comp.commonName.lower():
            matches.append(comp)
    return matches

# Find compounds containing 'chloride'
chloride_compounds = find_compound_by_name(doc, 'chloride')
print(f"Found {len(chloride_compounds)} compounds with 'chloride' in the name:")
for comp in chloride_compounds:
    print(f"  - {comp.commonName} (PubChem ID: {comp.pubChemID})")

# Find measurements at room temperature (298.15K)
room_temp_measurements = []
for fluid in doc.fluid:
    for measurement in fluid.sample.measurement:
        for param_value in measurement.parameterValue:
            if param_value.paramValue == 298.15:
                room_temp_measurements.append(measurement)
                break

print(f"\nFound {len(room_temp_measurements)} measurements at room temperature (298.15K)")
for measurement in room_temp_measurements:
    print(f"  - Measurement ID: {measurement.measurement_id}")
```
