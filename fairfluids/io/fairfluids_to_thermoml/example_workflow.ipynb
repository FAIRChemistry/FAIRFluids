# FAIRFluids -> ThermoML: Beispiel-Workflow

Dieses Notebook zeigt einen minimalen End-to-End-Run fuer den neuen Reverse-Konverter.

Pipeline:

`FAIRFluids JSON -> convert(...) -> ThermoML XML`

```python
from pathlib import Path
import json

from fairfluids.io.fairfluids_to_thermoml.converter import convert

output_dir = Path("./example_outputs")
output_dir.mkdir(parents=True, exist_ok=True)

fairfluids_doc = {
    "citation": {
        "title": "Reverse conversion demo",
        "doi": "10.1000/demo",
        "pub_name": "Demo Journal",
        "publication_year": "2026",
        "author": [{"family_name": "Mustermann", "given_name": "Max"}],
        "litType": "journal",
    },
    "compound": [
        {
            "compoundID": "compound_1",
            "commonName": "water",
            "standard_InChI": "InChI=1S/H2O/h1H2",
            "standard_InChI_key": "XLYOFNOQVPJJNP-UHFFFAOYSA-N",
        }
    ],
    "fluid": [
        {
            "fluidID": ["fluid_1"],
            "compounds": ["compound_1"],
            "property": [
                {
                    "propertyID": "property_1",
                    "properties": "density",
                    "unit": {"name": "kilogram per cubic meter"},
                }
            ],
            "parameter": [
                {
                    "parameterID": "parameter_1",
                    "parameters": "Temperature",
                    "unit": {"name": "kelvin"},
                },
                {
                    "parameterID": "parameter_2",
                    "parameters": "Pressure",
                    "unit": {"name": "kilopascal"},
                },
            ],
            "sample": {
                "measurement": [
                    {
                        "propertyValue": [
                            {
                                "propertyID": "property_1",
                                "propValue": 997.1,
                                "uncertainty": 0.2,
                            }
                        ],
                        "parameterValue": [
                            {"parameterID": "parameter_1", "paramValue": 298.15},
                            {"parameterID": "parameter_2", "paramValue": 101.325},
                        ],
                    },
                    {
                        "propertyValue": [
                            {"propertyID": "property_1", "propValue": 995.6}
                        ],
                        "parameterValue": [
                            {"parameterID": "parameter_1", "paramValue": 303.15},
                            {"parameterID": "parameter_2", "paramValue": 101.325},
                        ],
                    },
                ]
            },
        }
    ],
}

json_path = output_dir / "example_fairfluids.json"
json_path.write_text(json.dumps(fairfluids_doc, indent=2), encoding="utf-8")
json_path
```

```python
with open("/home/sga/Code/FAIRFluids/Workflows/outputs/Xu_Glycerol_Water.json", "r") as f:
    fairfluids_doc = json.load(f)
```

```python
xml_string = convert(fairfluids_doc)
xml_path = output_dir / "example_thermoml.xml"
xml_path.write_text(xml_string, encoding="utf-8")

print(f"ThermoML geschrieben nach: {xml_path}")
print(xml_string[:800])
```

```python
from fairfluids.io.thermoml_to_fairfluids.parser import parse

parsed = parse(xml_path)
print(f"Datensaetze: {len(parsed.datasets)}")
print(f"Compounds: {len(parsed.compounds)}")
print(f"Properties im ersten Datensatz: {len(parsed.datasets[0].properties)}")
print(f"Variablen im ersten Datensatz: {len(parsed.datasets[0].variables)}")
print(f"Constraints im ersten Datensatz: {len(parsed.datasets[0].constraints)}")
```
